```python
arr=np.zeros((3,4))
print(arr)
print(arr.ndim)
print(arr.shape)
```

```python
arr1=np.expand_dims(arr,axis=0) # increase the dimension by 1 in the provided axix,0 means rows(x axis) and 1 means cols (y axis)
print(arr1)
print(arr1.ndim)
print(arr1.shape)
```

```python
arr2=np.expand_dims(arr,axis=1) # increase the dimension by 1 in the provided axix,0 means rows(x axis) and 1 means cols (y axis)
print(arr2)
print(arr2.ndim)
print(arr2.shape)
```

```python
arr3=arr2.reshape((2,2,3))
print(arr3)
```

```python
arr4=np.expand_dims(arr3,axis=0)
print(arr4)
print(arr4.shape)
```

```python
arr5=np.expand_dims(arr3,axis=1)
print(arr5)
print(arr5.shape)
```

```python
# now to remove a dimension
arr3=np.squeeze(arr4,axis=0)
print(arr3)
print(arr3.shape)
```

```python
# now to remove a dimension
arr3=np.squeeze(arr5,axis=1)
print(arr3)
print(arr3.shape)
```
