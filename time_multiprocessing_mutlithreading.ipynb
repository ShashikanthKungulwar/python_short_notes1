```python
# iterable: DS that can return its members one by one example string, list etc abd the process is called iteration
#example List is iterable
L=[1,2,3,"ran"]
for i in L:
  print(i)


#iterator: is an onject represnting a steram of data. if any object can be converted to iter() then it is iterator (intermediate object)
a=iter(L)
print(a) # the obeject this is iterator for the list (intermediate object)

print(next(a))
```

```python
# following syntax will make exection time more and  memory usage also more to storage for large n values
def square(n):
  result=[]
  for i in range(n):
    result.append(i**2)
  return result
print(square(5))



# to avoidn this we can use generator function here one by one executoin is done instead of at a time as above
# here these kind of functions are called generator functions

def square_gen(n):
  for i in range(n):
    yield i**2
a=square_gen(5)
print(a)
print(next(a)) #similar to iterator concept
```

```python
# lambda, map, filter functions


# lambda functions
a=lambda x:x**2
print("lambda funciton output is ",a(5))


# map function
# syntax: map(func,*iterable)

b=map(square,[1,2,3,4,5])
print(list(b))
#we can directly pass lambda function too
c=map(lambda x:x**2,[1,2,3,4,5])
print(next(c))
print(list(c))
print(list(c))#once converted can use again

d=map(lambda x,y:x+y,[1,2,3,4,5],[2,4,5])
print(list(d))





#now reduce function it folds the iterable
from functools import reduce

e=reduce(lambda x,y:1 if(print(x) and 0) else x+y,[1,2,3,4,5],0)  #here in x acuumulated value gets stored for next itereation and 0 is initialzation for x done by 3rd parameter which is optional if not directly 1s element of iterable is choosen as initialization
print(e)





#now filter function it is used when we want to filter iterable based on any condition
f=filter(lambda x:x%2==0,[1,2,3,4,5])
print(list(f))



# in simple terms
# map is used to compute a funciton on each item of iterable
# reduce is to fold the iterable into 1
# filter is used to get a filterd iterable on any condition
```

```python
import time
import threading # it is a python module
flag=False
flag_lock=threading.Lock()

start=time.perf_counter()
def test_func(arg1,arg2):
  global flag
  print("blah blah blah")

  with flag_lock:
    flag=not flag
    print(f"sleep for {arg1 if flag else arg2} sec")
    time.sleep(arg1 if flag else arg2)
  print("done with sleeping")

threads=[]
for i in range(10):
  t=threading.Thread(target=test_func,args=[2,3])
  t.start()
  threads.append(t)

for i in threads:
  i.join()
  # pass

end=time.perf_counter()
print(f"the program finished in {round(end-start,2)} seconds") # here round function is rounding to 2 decimals
```

```python
import time
import threading


counter=0
counter_lock=threading.Lock()
start=time.perf_counter()


def increament_counter(thread):
  global counter
  with counter_lock:
    counter+=1
    print(f"thread {thread} increased counter to {counter}")
    time.sleep(1) # making counter locked for 1 sec
  #time.sleep(1) no use because we can keep lock as counter_lock is ended on before line


threads=[threading.Thread(target=increament_counter,args=[i]) for i in range(10)]

for i in threads:
  i.start()
for i in threads:
  i.join()

# for i in threads:
#   i.start()                 here joining before starting of next thread so ot will make serial execution
#   i.join()

end=time.perf_counter()
print(f"the program finished in {round(end-start,2)} seconds")
```

```python
import multiprocessing


def enroll_student(student_q):
    students = ["rahul", "raj", "man"]
    for i in students:
        student_q.put(i)
    student_q.put(None)   # termination signal
    print("Producer finished")


def register_student(student_q):
    while True:
        student = student_q.get()
        if student is None:
            print("Consumer received termination")
            break
        print(f"haii registering data of : {student}")


# if __name__ == "__main__":
student_q = multiprocessing.Queue()
p1 = multiprocessing.Process(target=enroll_student, args=(student_q,))
p2 = multiprocessing.Process(target=register_student, args=(student_q,))

p1.start()
p2.start()

p1.join()
p2.join()
print("Both processes finished")
```
