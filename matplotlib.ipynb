```python
#pie chart: it is used to understand the percent distributoin of data
langs=["e",'s','m']
students=[10,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%")
plt.show()
```

```python
#pie chart: it is used to understand the percent distributoin of data
langs=["e",'s','m']
explode=[0.0,0.1,0.1]
students=[10,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%",explode=explode)
plt.show()
```

```python
#pie chart: it is used to understand the percent distributoin of data
langs=["e",'s','m']
explode=[0.0,0.1,0.1]
students=[10,100,40]
plt.pie(students,labels=langs,autopct="%1.1f%%",explode=explode,shadow=True)
plt.show()
```

```python
# area plot
years=[2021,2018,2022,2024,2017]
cat1=[20,25,12,5,3]
cat2=[20,5,12,25,30]
cat3=[20,2,1,55,3]


plt.stackplot(years,cat1,cat2,cat3,labels=["cat1","cat2","cat3"],colors=["blue","orange","green"])
plt.xlabel("years")
plt.ylabel("Values")
plt.title("Area plot")
plt.legend()
plt.show()
```
